```python
import openai
from gtts import gTTS
import os

# Установи свой API-ключ от OpenAI
openai.api_key = 'твой_api_ключ'

# Функция для генерации текста с помощью GPT
def generate_text(prompt):
    response = openai.Completion.create(
        engine="text-davinci-003",  # Модель GPT
        prompt=prompt,  # Текст-запрос
        max_tokens=150,  # Максимальное количество токенов в ответе
        temperature=0.7,  # Креативность (от 0 до 1)
    )
    return response.choices[0].text.strip()

# Функция для озвучки текста
def text_to_speech(text, language='ru'):
    tts = gTTS(text=text, lang=language)
    tts.save("output.mp3")
    os.system("start output.mp3")  # Для Windows
    # os.system("afplay output.mp3")  # Для macOS
    # os.system("mpg321 output.mp3")  # Для Linux

# Основной код
if __name__ == "__main__":
    # Запрос для генерации текста
    prompt = "Напиши короткий рассказ про космос."

    # Генерация текста
    generated_text = generate_text(prompt)
    print("Сгенерированный текст:")
    print(generated_text)

    # Озвучка текста
    text_to_speech(generated_text)
```

```python
# Установка зависимостей
!pip install datasets transformers torch soundfile

# Импорт библиотек
from transformers import SpeechT5Processor, SpeechT5ForTextToSpeech
from datasets import load_dataset
import torch
import soundfile as sf
import numpy as np
from huggingface_hub import login
from google.colab import userdata
from IPython.display import Audio

# Получение токена из секретов Colab
try:
    hf_token = userdata.get('HF_TOKEN')  # Убедитесь, что имя совпадает с именем в секретах
    login(hf_token)
    print("Аутентификация успешна!")
except Exception as e:
    print(f"Ошибка при аутентификации: {e}")

# Загрузка модели и процессора
processor = SpeechT5Processor.from_pretrained("microsoft/speecht5_tts")
model = SpeechT5ForTextToSpeech.from_pretrained("microsoft/speecht5_tts")

# Загрузка speaker embeddings (например, из датасета cmu-arctic-xvectors)
dataset = load_dataset("Matthijs/cmu-arctic-xvectors", split="validation")
speaker_embeddings = torch.tensor(dataset[0]["xvector"]).unsqueeze(0)
print("Форма speaker_embeddings:", speaker_embeddings.shape)
print("Пример значений:", speaker_embeddings[0][:10])  # Первые 10 значений

# Генерация текста
input_text = "Привет, как дела?"
print("Входной текст:", input_text)

# Токенизация текста
inputs = processor(text=input_text, return_tensors="pt")
print("Токенизированный текст:", inputs)

# Генерация речи с использованием speaker embeddings
with torch.no_grad():
    speech = model.generate_speech(inputs["input_ids"], speaker_embeddings)
    print("Форма аудио:", speech.shape)
    print("Тип данных аудио:", speech.dtype)

# Увеличение громкости
speech = speech.numpy()  # Преобразуем в numpy array
speech = speech * 10.0   # Увеличиваем громкость в 10 раз (можно настроить)
speech = np.clip(speech, -1.0, 1.0)  # Ограничиваем значения, чтобы избежать искажений

# Сохранение аудио в файл
output_file = "output.wav"
sf.write(output_file, speech, samplerate=22050)
print(f"Аудио сохранено в файл: {output_file}")

# Воспроизведение аудио
Audio(output_file)

# Скачивание файла (опционально)
from google.colab import files
files.download(output_file)
```
