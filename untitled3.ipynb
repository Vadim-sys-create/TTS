```python
import streamlit as st
from transformers import SpeechT5Processor, SpeechT5ForTextToSpeech
from datasets import load_dataset
import torch
import soundfile as sf
import numpy as np

# Заголовок приложения
st.title("Генерация речи с помощью SpeechT5")

# Загрузка модели и процессора
@st.cache_resource
def load_model():
    processor = SpeechT5Processor.from_pretrained("microsoft/speecht5_tts")
    model = SpeechT5ForTextToSpeech.from_pretrained("microsoft/speecht5_tts")
    return processor, model

processor, model = load_model()

# Загрузка speaker embeddings
@st.cache_resource
def load_speaker_embeddings():
    dataset = load_dataset("Matthijs/cmu-arctic-xvectors", split="validation")
    speaker_embeddings = torch.tensor(dataset[0]["xvector"]).unsqueeze(0)
    return speaker_embeddings

speaker_embeddings = load_speaker_embeddings()

# Ввод текста
input_text = st.text_area("Введите текст для озвучки:", "Привет, как дела?")

# Генерация речи
if st.button("Озвучить"):
    inputs = processor(text=input_text, return_tensors="pt")
    with torch.no_grad():
        speech = model.generate_speech(inputs["input_ids"], speaker_embeddings)
        speech = speech.numpy() * 10.0  # Увеличение громкости
        speech = np.clip(speech, -1.0, 1.0)

    # Сохранение аудио
    output_file = "output.wav"
    sf.write(output_file, speech, samplerate=22050)

    # Воспроизведение аудио
    st.audio(output_file, format="audio/wav")

    # Скачивание аудио
    with open(output_file, "rb") as f:
        st.download_button("Скачать аудио", f, file_name="output.wav")
```

```python
from telegram import Update
from telegram.ext import ApplicationBuilder, CommandHandler, MessageHandler, filters
from transformers import SpeechT5Processor, SpeechT5ForTextToSpeech
from datasets import load_dataset
import torch
import soundfile as sf
import numpy as np
import os

# Загрузка модели и процессора
processor = SpeechT5Processor.from_pretrained("microsoft/speecht5_tts")
model = SpeechT5ForTextToSpeech.from_pretrained("microsoft/speecht5_tts")

# Загрузка speaker embeddings
dataset = load_dataset("Matthijs/cmu-arctic-xvectors", split="validation")
speaker_embeddings = torch.tensor(dataset[0]["xvector"]).unsqueeze(0)

# Функция для генерации речи
def generate_speech(text):
    inputs = processor(text=text, return_tensors="pt")
    with torch.no_grad():
        speech = model.generate_speech(inputs["input_ids"], speaker_embeddings)
        speech = speech.numpy() * 10.0  # Увеличение громкости
        speech = np.clip(speech, -1.0, 1.0)
    output_file = "output.wav"
    sf.write(output_file, speech, samplerate=22050)
    return output_file

# Обработчик команды /start
async def start(update: Update, context):
    await update.message.reply_text("Привет! Отправь мне текст, и я озвучу его.")

# Обработчик текстовых сообщений
async def text_to_speech(update: Update, context):
    text = update.message.text
    audio_file = generate_speech(text)
    await update.message.reply_audio(audio=open(audio_file, "rb"))
    os.remove(audio_file)  # Удаляем файл после отправки

# Запуск бота
if __name__ == "__main__":
    token = os.getenv("7726156630:AAFaqW_KgTawqEz2ikkUopk9MJTrvyNf1Io")  # Токен бота из переменных окружения
    application = ApplicationBuilder().token(token).build()

    application.add_handler(CommandHandler("start", start))
    application.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, text_to_speech))

    application.run_polling()
```
